```python
%pylab inline
import scipy.stats as stats
import seaborn as sns
sns.set(style='ticks')
```

```python
# Standard normal distribtuion, stats.norm(mu,sigma)
rv=stats.norm(0,1)
x=r_[-5:5:0.1]
plot(x,rv.pdf(x))
```

```python
# shiffed distribution
x=arange(-5,15,0.1)
#Plot a normal distribution
myMean=5
mySD=2
y=stats.norm.pdf(x,myMean, mySD)
plot(x,y)
title('Shifted Normal Distribution')
```

```python
#Random numbers with a normal distribution
numData=500
data=stats.norm.rvs(myMean, mySD, size=numData)
plot(data,'.')
title('Normally distributed data')
show()
```

```python
hist(data)
title('Histogram of normally distributed data')
```

```python
# multiple normal sample distributuions
'''Show multiple samples from the same distribution, and compared means'''
# Do this 25 times, and show the histograms
numRows=5
numData=100
for ii in range(numRows):
    for jj in range(numRows):
        data=stats.norm.rvs(myMean, mySD, size=numData)
        subplot(numRows, numRows, numRows*ii+jj+1)
        hist(data)
        
        xticks([])
        yticks([])
        xlim(myMean-3*mySD, myMean+3*mySD)
tight_layout()
show()
```

```python
# Check out mean of 1000 normally distributed samples
numTrials=1000
numData=100
myMeans=ones(numTrials)*nan
for ii in range(numTrials):
    data=stats.norm.rvs(myMean, mySD, size=numData)
    myMeans[ii]=mean(data)
print('The standard error of the mean, with {0} samples, is {1:5.3f}'
      .format(numData,std(myMeans)))
```

```python
# Normality Check
'''Check if the distribution is normal.'''
# Generate and show a distribution
numData=100
data=stats.norm.rvs(myMean, mySD, size=numData)
hist(data)
```

```python
# Graphical test
checkNorm=stats.probplot(data, plot=plt)
# Calculate quantiles for a probability plot
```

```python
_, pVal=stats.normaltest(data)
if pVal>0.05:
    print('Data are probably normally distributed')
```
