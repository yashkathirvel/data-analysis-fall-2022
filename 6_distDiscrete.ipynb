```python
%pylab inline
from scipy import stats
import numpy as np
```

```python
# Bernoulli Distribtuion
p=0.5
bernoulliDist=stats.bernoulli(p)
p_tails=bernoulliDist.pmf(0)
p_heads=bernoulliDist.pmf(1)
```

```python
# Simulate 10 Bernoulli trails
trails=bernoulliDist.rvs(10)
trails
```

```python
# Binomial distribution, stats.binom(n,p)
bd1=stats.binom(20,0.5)
bd2=stats.binom(20,0.7)
bd3=stats.binom(40,0.5)
k=arange(40)
plot(k,bd1.pmf(k),'o-b')
plot(k,bd2.pmf(k),'d-r')
plot(k,bd3.pmf(k),'s-g')
title('Binomial Distribution')
legend(['p=0.5 and n=20','p=0.7 and n=20','p=0.5 and n=40'])
xlabel('X')
ylabel('P(X)')
```

```python
# Calculate the probabilities how often heads come up during four tosses
(p,num)=(0.5,4)
binomDist=stats.binom(num,p)
binomDist.pmf(arange(5))
```
